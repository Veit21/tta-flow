# This notebook creates data configurations for training/testing a network

```python
# Libraries
import os

import pandas as pd

from natsort import natsorted
from pathlib import Path
```

```python
## Paths
data_root = Path(os.getenv("DATA_ROOT", "/app/data/retouch"))               
out_root  = Path(os.getenv("OUT_ROOT", "/app/data/dataset_configs"))
spectralis  = natsorted(data_root.rglob("spectralis/*.npy"))
cirrus      = natsorted(data_root.rglob("cirrus/*.npy"))
topcon      = natsorted(data_root.rglob("topcon/*.npy"))

# Volumes
vol_spectralis  = [pth for pth in spectralis if "annotations" not in str(pth)]
vol_cirrus      = [pth for pth in cirrus if "annotations" not in str(pth)]
vol_topcon      = [pth for pth in topcon if "annotations" not in str(pth)]
print(f"Found\n{len(vol_spectralis)} Spectralis volumes\n{len(vol_cirrus)} Cirrus volumes\n{len(vol_topcon)} Topcon volumes")

# Annotations
mask_spectralis  = [pth for pth in spectralis if "annotations" in str(pth)]
mask_cirrus      = [pth for pth in cirrus if "annotations" in str(pth)]
mask_topcon      = [pth for pth in topcon if "annotations" in str(pth)]
print(f"Found\n{len(mask_spectralis)} Spectralis annotations\n{len(mask_cirrus)} Cirrus annotations\n{len(mask_topcon)} Topcon annotations")
```

```python
# Create dataframes
df_spectralis = pd.DataFrame({"volume": vol_spectralis, "mask": mask_spectralis})
df_cirrus     = pd.DataFrame({"volume": vol_cirrus, "mask": mask_cirrus})
df_topcon     = pd.DataFrame({"volume": vol_topcon, "mask": mask_topcon})
print("DataFrames created")
```

```python
# Save dataframes as csv
df_spectralis.to_csv(out_root / "retouch_original_spectralis.csv", index=False)
df_cirrus.to_csv(out_root / "retouch_original_cirrus.csv", index=False)
df_topcon.to_csv(out_root / "retouch_original_topcon.csv", index=False)
print(f"DataFrames saved to {out_root}")
```
